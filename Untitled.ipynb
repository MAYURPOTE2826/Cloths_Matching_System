```python
import random
import pandas as pd
import numpy as np
colors = {
    "red": "#FF0000",
    "black": "#000000",
    "white": "#FFFFFF",
    "blue": "#0000FF",
    "navy": "#000080",
    "beige": "#F5F5DC",
    "grey": "#808080",
    "green": "#008000",
    "brown": "#A52A2A",
    "cream": "#FFFDD0"
}

# Realistic shirt → pant mappings
matching_rules = {
    "red": ["black", "white", "beige"],
    "blue": ["grey", "white", "beige"],
    "black": ["white", "cream", "grey"],
    "white": ["black", "blue", "brown"],
    "grey": ["black", "navy", "white"],
    "green": ["beige", "white", "brown"],
    "brown": ["cream", "white"],
    "navy": ["grey", "white"],
    "beige": ["brown", "black"],
    "cream": ["brown", "black"]
}

rows = []
for shirt_color, pant_list in matching_rules.items():
    for pant_color in pant_list:
        for _ in range(15):  # multiple samples for balance
            rows.append({
                "shirt_color_name": shirt_color,
                "shirt_hex": colors[shirt_color],
                "pant_color_name": pant_color,
                "pant_hex": colors[pant_color],
                "accessory_suggestion": random.choice(["watch", "belt", "sunglass", "cap"]),
                "style_type": random.choice(["formal", "casual"])
            })

rows = []
for shirt_color, pant_list in matching_rules.items():
    for pant_color in pant_list:
        for _ in range(40):  # increase from 15 to 40
            rows.append({
                "shirt_color_name": shirt_color,
                "shirt_hex": colors[shirt_color],
                "pant_color_name": pant_color,
                "pant_hex": colors[pant_color],
                "accessory_suggestion": random.choice(["watch", "belt", "sunglass", "cap", "bracelet"]),
                "style_type": random.choice(["formal", "casual", "party", "sport"])
            })


df = pd.DataFrame(rows)
df.to_csv("fashion_color_dataset_refined.csv", index=False)
print("✅ Dataset Created:", len(df), "samples")
df.head()
```
